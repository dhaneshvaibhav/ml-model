```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('//content/spam-email-classification/email.csv')
data.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.shape
```

```python
data.loc[data['Category']=='spam','Category',]=0
data.loc[data['Category']=='ham','Category',]=1
```

```python
x=data['Message']
y=data['Category']
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x1,x2,y1,y2 = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
x1.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics import accuracy_score
```

```python
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x1_features = feature_extraction.fit_transform(x1)
x2_features = feature_extraction.transform(x2)
y1 = y1.fillna(0).astype('int') # Replace NaN values with 0 and convert to integer
y2 = y2.fillna(0).astype('int')
```

```python
print(x1)
```

```python
print(x1_features)
```

```python
model=LogisticRegression()
model.fit(x1_features,y1)
```

```python
pred=model.predict(x1_features)
accs=accuracy_score(y1,pred)
```

```python
print(accs)
```

```python
predictio_test=model.predict(x2_features)
accuracy_test=accuracy_score(y2,predictio_test)
```

```python
print(accuracy_test)
```
